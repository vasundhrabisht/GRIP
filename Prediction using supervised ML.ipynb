# Task 1 : Prediction using Supe

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression() #creating a linear regression object
```

```python
import pandas as pd
```

```python
df=pd.read_csv("http://bit.ly/w-data") #reading the data into a Pandas dataframe
```

```python
df
```

```python
x=df[["Hours"]]
y=df["Scores"]
```

```python
lm.fit(x,y)
```

### A Linear Relationship between predictor variable (x) and targer variable (y) is depicted by : y = a + bx where a : intercept and b : slope or coefficient of regression

```python
lm.intercept_ #value of the intercept i.e. a
```

```python
lm.coef_ #value of the coefficient i.e. b
```

```python
import seaborn as sns
```

```python
sns.regplot(x,y)  #Regression Plot
```

### Hence we observe a positive relationship between no of hours and percentage

### Making Predictions :-

```python
x_value=[[9.25]]
```

```python
yhat=lm.predict(x_value)
```

```python
print("The predicted score at ",x_value," hours is ",yhat)
```
