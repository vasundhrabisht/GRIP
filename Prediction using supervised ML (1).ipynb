## Task 1 : Predicting score from number of hours

### Importing required libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

### Importing data into a pandas dataframe 

```python
df=pd.read_csv("http://bit.ly/w-data") 
```

```python
df.head()
```

### Splitting data into training and testing sets.

```python
x=df[["Hours"]]
y=df["Scores"]
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=0)
```

### Building and training the model with training set 

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

### A Linear Relationship between predictor variable (x) and targer variable (y) is depicted by : y = a + bx where a : intercept and b : slope or coefficient of regression

```python
lm.intercept_ #value of the intercept i.e. a
```

```python
lm.coef_ #value of the coefficient i.e. b
```

### Plotting the obtained regression line

```python
line = lm.coef_*x+lm.intercept_
plt.scatter(x, y)
plt.plot(x, line);
plt.show()
```

### Hence we observe a positive relationship between no of hours and percentage

### Using testing set to assess the performance

```python
yhat=lm.predict(x_test)
assessment = pd.DataFrame({'Actual Value': y_test,'Predicted Value': yhat})
assessment
```

### Predicting percentage for 9.25 hours 

```python
x_value=[[9.25]]
y_predicted= lm.predict(x_value)
print("The predicted score at ",x_value," hours is ",y_predicted)
```

### Measure of in-sample evaluation

```python
lm.score(x,y)
```

### This is coefficient of determination or R^2. Since its value is close to 1 we conclude that the line is a good fit for the data.
